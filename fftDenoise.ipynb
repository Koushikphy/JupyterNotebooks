## Remove noise from data using Fourier Transform

```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
n = 256     
```

```python
# add the x axis
x = np.linspace(-np.pi,np.pi,n) 
```

```python
# y values, simple sin function but added with some noise
y = np.sin(x)+0.1*np.random.rand(n) 
```

```python
# lets plot the data 
plt.plot(x,y)
```

```python
# now get the amplitudes of frequencies using fourier transform 
yHat = np.fft.fft(y)
amp = np.real(yHat*np.conj(yHat))
```

```python
plt.plot(x,amp)
plt.ylim(0,100)
```

```python
# denoise the data by making the amplitudes for the noise frequencies 0
yHat[amp<10]=0
```

```python
# Get the de noised data by transforming back to real spacs
z = np.fft.ifft(yHat)
plt.plot(x,z.real)
```
